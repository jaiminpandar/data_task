```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
data=pd.read_csv('train.csv')
data.head()
```

```python
feature=data[['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
feature=feature.dropna()
```

```python
feature.head()
```

```python
feature['Sex'].replace('male', 0,inplace=True)
feature['Sex'].replace('female', 1,inplace=True)
```

```python
feature.head()
```

```python
feature['Embarked'].replace('S', 0,inplace=True)
feature['Embarked'].replace('C', 1,inplace=True)
feature['Embarked'].replace('Q', 2,inplace=True)
```

```python
feature.head()
```

```python
x=feature.drop('Survived',axis=1)
x
```

```python
y=feature['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
y_test
```

```python
model=DecisionTreeClassifier(max_depth=3)
model.fit(x_train,y_train)
print(model)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
a=metrics.accuracy_score(y_pred,y_test)
a
```
